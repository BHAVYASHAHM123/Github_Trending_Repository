```python
import bs4
import requests
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
a = "https://github.com/"
b = "trending"
res = a+b
print(res)
```

```python
res = requests.get(res)
```

```python
res.text
```

```python
soup = bs4.BeautifulSoup(res.text, 'lxml')
```

```python
soup
```

```python
# Scraping the name of the Repository description

descriptions = []
for description in soup.find_all('p',  class_ = 'col-9 color-fg-muted my-1 pr-4'):
    descriptions.append(description.text)
    
descriptions = descriptions[:20]
print(descriptions)
```

```python
# Using the strip function we can remove the \n value from the list  in discreption
Descriptions = []

for convetring_description in descriptions:
    Descriptions.append(convetring_description.strip())

print(Descriptions)
```

```python
# Scraping the number of contribution of the Repository .

contributions = []
for contribution in soup.find_all('a',  class_ = 'Link--muted d-inline-block mr-3'):
    contributions.append(contribution.text)
    
print(contributions)
```

```python
# Using the strip function we can remove the \n value from the list  in Contribution 

Contributions = []

for Converting_Contribution in contributions:
    Contributions.append(Converting_Contribution.strip())
Contributions = Contributions[:20]
print(Contributions)
```

```python
# Scraping the number of language of the Repository .

languages = []
for language in soup.find_all('span',  itemprop = 'programmingLanguage'):
    languages.append(language.text)
    
languages = languages[:20]
print(languages)
```

```python
# Scraping the number of name of the Repository .

names = []
for name in soup.find_all('h1', class_ = "h3 lh-condensed" ):
    names.append(name.text)
    
names = names[:20]
print(names)
```

```python
# Using the strip function we can remove the \n value from the list  in names 
Names = []

for convetring_name in names:
    Names.append(convetring_name.strip())

print(Names)
```

```python
lst=Names
names = []
for x in lst:
    names.append(x.replace("\n", ""))
print(list(names))
```
